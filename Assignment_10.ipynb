<b>1. How do you distinguish between shutil.copy() and shutil.copytree()?

shutil.copy() copies a file to destination. If the destination is a directory then the file is copied if directory is present. It returns error if the directory doesn't exists. shutil.copytree() recursively copies the files and sub-directories in a specified directory tree to specified location. It creates a directory if destination directory is not already present.

<b>5. Create a programme that searches a folder tree for files with a certain file extension (such as .pdf
or .jpg). Copy these files from whatever location they are in to a new folder.

```python
import os
import shutil

def copy_matched_files(src, dest, match_extension):
    if os.path.exists(src) == False:
        raise FileNotFoundError(f"{src} doesn't exist")
    if os.path.exists(dest):
        shutil.rmtree(dest)
    src_files_matched = []
    dest_files_loc = []
    os_walk = os.walk(src)
    for dir, _, file_list in os_walk:
        src_files_matched.extend([os.path.join(dir, file) for file in file_list if file.endswith(match_extension)])
        dest_dir = os.path.join(dest, os.path.sep.join(dir.split(os.path.sep)[1:]))
        dest_files_loc.extend([os.path.join(dest_dir, file) for file in file_list if file.endswith(match_extension)])
        
    for src_file, dest_file in zip(src_files_matched, dest_files_loc):
        dest_dir = os.path.dirname(dest_file)
        if os.path.exists(dest_dir) == False:
               os.makedirs(dest_dir)
        shutil.copy(src_file, dest_file)
        
copy_matched_files('os_assignment', 'os_test', '.txt')
```

```python
src_walk = os.walk('os_assignment')
dest_walk = os.walk('os_test')
print(f"{'-' * 30}\n'os_assignment' folder structure\n{'-' * 30}")
for dir, _, file in src_walk:
    file_hierarchy = "\n\t-".join(file)
    print(f'{dir}\n\t-{file_hierarchy}')
    
print(f"\n{'-' * 30}\n'os_test' folder structure\n{'-' * 30}")
for dir, _, file in dest_walk:
    file_hierarchy = "\n\t-".join(file)
    print(f'{dir}\n\t-{file_hierarchy}')
```

The source and destination folder hierarchy shows .txt files from 'os_assignment' are copied to 'os_test' folder in the respective sub-folders.
